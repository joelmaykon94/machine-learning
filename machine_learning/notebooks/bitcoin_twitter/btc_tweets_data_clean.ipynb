```python
def dt(x):
    t = pd.Timestamp(x)
    return pd.Timestamp.date(t)

df.timestamp = df.timestamp.apply(dt)
```

```python
df_price_btc = pd.read_csv('../data/raw/BitcoinPriceHistory.csv')
```

```python
df_price_btc.Date = df_price_btc.Date.apply(dt)
```

```python
df_price_btc.head()
```

```python
df_price_btc.tail()
```

```python
df_price_btc['volume'] = df_price_btc["Vol."]
df_price_btc['variance_price'] = df_price_btc["Change %"]
```

```python
df_price_btc = df_price_btc.drop(["Vol.", "Change %"],  axis=1, inplace=False)
```

```python
def convert_to_float(value):
    try:
        if "K" in value.upper():
            return float(value.replace("K", ""))
        elif "%" in value.upper():
            return float(value.replace("%", ""))
        else:
            return float(value.replace(",", ""))

    except (ValueError, AttributeError):
        # Handle potential conversion errors or missing values
        return 0
```

```python
#df_price_btc.apply(convert_to_float)
# Apply the function to 'col1'
df_price_btc["Price"] = df_price_btc["Price"].apply(convert_to_float)
df_price_btc["Open"] = df_price_btc["Open"].apply(convert_to_float)
df_price_btc["High"] = df_price_btc["High"].apply(convert_to_float)
df_price_btc["Low"] = df_price_btc["Low"].apply(convert_to_float)
df_price_btc["volume"] = df_price_btc["volume"].apply(convert_to_float)
df_price_btc["variance_price"] = df_price_btc["variance_price"].apply(convert_to_float)
df_price_btc.head()
```

```python
df_price_btc.info()
```

```python
#criando campo de data nos dois datasets antes de uni-los
df['date'] = df.timestamp
df_price_btc['date'] = df_price_btc.Date
# Definindo a coluna de chave prim√°ria (assumindo que ambas possuem 'data')
key = 'date'
# Realizando a jun√ß√£o interna
df_origin = df_price_btc.merge(df, on=key, how='inner')

df_origin.head()
```

```python
df_origin = df_origin.drop(['Date', 'timestamp'], axis=1, inplace=False)
```

```python

df_origin.columns = df_origin.columns.str.lower()

df_origin.head()
```

```python
df_origin.head()
```

```python
text = list(df_origin.loc[:,'text\r'].values)
```

```python
from transformers import AutoTokenizer ,pipeline
from transformers import TFAutoModelForSequenceClassification

finbert = TFAutoModelForSequenceClassification.from_pretrained("../models/ProsusAI-finbert")
tokenizer = AutoTokenizer.from_pretrained("../models/ProsusAI-finbert")
classifier = pipeline('sentiment-analysis', model=finbert, tokenizer=tokenizer)
```

```python
label_and_score = classifier(text)
labels =[pred['label'] for pred in label_and_score]
labels = list(map(lambda x: x.replace('positive','1'), labels))
labels = list(map(lambda x: x.replace('neutral','0'), labels))
labels = list(map(lambda x: x.replace('negative','-1'), labels))
labels = list(map(int, labels))
df_origin.loc[:,'label'] = labels
```
